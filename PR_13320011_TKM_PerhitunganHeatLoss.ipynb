### Data Soal

```python
#Konversi Satuan Inch ke Meter
def InchtoM(inch):
    return inch*0.0254

# Properti 
# Temperature Steam dalam Kelvin
Ti = (180 + 1+1) + 273

# Temperature Lingkungan dalam Kelvin
Tenv = 30 + 1 + 273

# Kecepatan udara melintang
vudara = 2


# Properti Pipa
# Pipa carbonsteel dengan diameter 12in (schedule 40)
# Referensi : https://sinarindo.co.id/index.php/product/77-pipes/75-schedule-40
# OD : 318.5 mm
# ID : 297.9 mm 
r1 = (297.9/2 * 10**(-3))
r2 = ((318.5/2) * 10**(-3))
# Properti Carbonsteel Plain Carbon Tabel A1
Tpipa = [300, 400, 600, 800, 1000]
kpipa = [60.5, 56.7, 48.0, 39.2, 30.0]
Dpipa = InchtoM(12)


# Properti Insulator
# Tebal : 1 inch
r3 = r2 + InchtoM(1)
# Nilai konduktivitas termal
kinsulasi = 0.03

# Properti Clading
# Tebal : 2 mm 
r4 = r3 + 2 * 10**(-3)
kcladding = 52

#Tebal Pipa dengan Insulasi
Dinsulasi = Dpipa + r4

# Properti Fluida Data Tabel A4
TA4 = [100,	150,	200,	250,	300,	350,	400,	450,	500,	550,	600,	650,	700,	750,	800,	850,	900,	950,	1000,	1100,	1200,	1300,	1400,	1500,	1600,	1700,	1800,	1900,	2000,	2100,	2200,	2300,	2400,	2500,	3000]
rho = [3.5562, 2.3364, 1.7458,	1.3947,	1.1614,	0.9950,	0.8711,	0.7740,	0.6964,	0.6329,	0.5804,	0.5356,	0.4975,	0.4643,	0.4354,	0.4097,	0.3868,	0.3666,	0.3482,	0.3166,	0.2902,	0.2679,	0.2488,	0.2322,	0.2177,	0.2049,	0.1935,	0.1833,	0.1741,	0.1658,	0.1582,	0.1513,	0.1448,	0.1389,	0.1135]
miu = [71.1,	103.4,	132.5,	159.6,	184.6,	208.2,	230.1,	250.7,	270.1,	288.4,	305.8,	322.5,	338.8,	354.6,	369.8,	384.3,	398.1,	411.3,	424.4,	449.0,	473.0,	496.0,	530,	557,	584,	611,	637,	663,	689,	715,	740,	766,	792,	818,	955]
kf = [9.34,	13.8,	18.1,	22.3,	26.3,	30.0,	33.8,	37.3,	40.7,	43.9,	46.9,	49.7,	52.4,	54.9,	57.3,	59.6,	62.0,	64.3,	66.7,	71.5,	76.3,	82,	91,	100,	106,	113,	120,	128,	137,	147,	160,	175,	196,	222,	486]
Pr = [0.786,	0.758,	0.737,	0.720,	0.707,	0.700,	0.690,	0.686,	0.684,	0.683,	0.685,	0.690,	0.695,	0.702,	0.709,	0.716,	0.720,	0.723,	0.726,	0.728,	0.728,	0.719,	0.703,	0.685,	0.688,	0.685,	0.683,	0.677,	0.672,	0.667,	0.655,	0.647,	0.630,	0.613,	0.536]
```

### Regresi Property

```python
import numpy as np
import matplotlib.pyplot as plt

Npoly = 12
def regresi(x, data, y, title):
    plt.figure(figsize = (8,6))
    plt.xlabel('T', size = 14)
    plt.ylabel(y, size = 14)
    plt.scatter(x,data, label = 'data')
    plt.title(title)
    plt.grid('True')
    RegEquation = np.poly1d(np.polyfit(x, data, Npoly))
    print(RegEquation)
    plt.plot(x, RegEquation(x), color = 'r')
    plt.show()
    return RegEquation



# Regresi Data Tabel A1
# Regresi data temperature terhadap konduktivitas pipa berdasar Tabel A1
Regkpipa = regresi(Tpipa,kpipa,"kpipa","Regresi Temperature dengan kpipa - Tabel A1")

# Regresi Data Tabel A4
# Membuat regresi data temperature terhadap massa jenis udara berdasar Tabel A4
Regrho = regresi(TA4, rho, "rho","Regresi Temperature terhadap Massa Jenis Udara - Tabel A4" )

# Regresi data temperature terhadap viskositas dinamik fluida udara berdasar Tabel A4
Regmiu = regresi(TA4, np.multiply(miu,10**(-7)) , 'miu', "Regresi Temperature terhadap Viskositas Dinamik Udara - Tabel A4")

# Regresi data temperature terhadap konduktivitas udara berdasar Tabel A4
Regkf = regresi(TA4, np.multiply(kf,10**(-3)) , 'kf', "Regresi Temperature terhadap Konduktivitas Udara - Tabel A4")

# Regresi data temperature terhadap bilangan Prandault udara berdasar Tabel A4
RegPr = regresi(TA4, np.multiply(Pr,10**(-3)) , 'Pr', "Regresi Temperature terhadap Bilangan Prandault Udara - Tabel A4")



```

### Perhitungan Tanpa Insulator

```python

#Tebakan awal 
Tf = Tenv + 1 #Temperature Film
toleransi = 0.00001

error = 1
iter= 1

# Fungsi menghitung Resistansi konduktif silinder
def conductresistcirc(rin,rout,k):
    return (np.log(rout/rin))/(2*np.pi*k)

# Fungsi menghitung Reynold Number
def numberRe(rho,v,D,miu):
    return (rho*v*D)/(miu) 

# Fungsi menghitung Nusselt Number dengan pendekatan Churchill and Bersnstein Equation
def numberNu(Re,Pr):
    return  0.3 + ((0.62*(Re**(1/2))*(Pr**(1/3))) / ((1+(0.4/Pr)**(2/3))**(1/4))) * ((1+(Re/282000)**(5/6))**(4/5))

def hcoef(Nu,kf,D):
    return (Nu*kf)/D

# Fungsi menghitung heatloss dengan pendekatan resistansi
def calculateq(Ti,Tenv,Rtot):
    return (Ti-Tenv)/Rtot 


heatloss = [0]

print(f'Iterasi\t Re\t\tPr\t\tNu\t\th\t\tq\t\tTs\t\tTf\t\tError')

while error > toleransi:
    
    # Kasus Tanpa Insulator

    # Bilangan Reynold
    Re = numberRe(Regrho(Tf),vudara,Dpipa,Regmiu(Tf)) 
    Nu = numberNu(Re,RegPr(Tf))
    h = hcoef(Nu,Regkf(Tf),Dpipa)


    # Perhitungan Resistansi
    # Inisiasi array dinamik resistansi
    resistansi = [0]
    resistansi.append(conductresistcirc(r1,r2,Regkpipa(Ti)))
    resistansi.append((1/(2*np.pi*h*r2)))

    resistansiTotal = np.sum(resistansi)

    heatloss.append(calculateq(Ti,Tenv,resistansiTotal))
    
    Ts = heatloss[1]*resistansi[len(resistansi)-1]+Tenv
    Tf2 = (Ts+Tenv)/2
    
    error = abs(((Tf2-Tf)/Tf)*100)

    print(f'{iter} \t {Re:.4f} \t {RegPr(Tf):.4f} \t {Nu:.4f} \t {h:.4f} \t {heatloss[1]:.4f} \t {Ts-273:.4f} \t {Tf-273:.4f} \t {error:.4f}')

    Tf = Tf2
    iter = iter +1

print("Heat loss per satuan panjang dari pipa adalah sebesar :",heatloss[1], "Watt/m")

```

### Perhitungan Dengan Insulator

```python

#Tebakan awal 
Tf = Tenv + 1 #Temperature Film
toleransi = 0.00001

error = 1
iter= 1

# Fungsi menghitung Resistansi konduktif silinder
def conductresistcirc(rin,rout,k):
    return (np.log(rout/rin))/(2*np.pi*k)

# Fungsi menghitung Reynold Number
def numberRe(rho,v,D,miu):
    return (rho*v*D)/(miu) 

# Fungsi menghitung Nusselt Number dengan pendekatan Churchill and Bersnstein Equation
def numberNu(Re,Pr):
    return  0.3 + ((0.62*(Re**(1/2))*(Pr**(1/3))) / ((1+(0.4/Pr)**(2/3))**(1/4))) * ((1+(Re/282000)**(5/6))**(4/5))

def hcoef(Nu,kf,D):
    return (Nu*kf)/D

# Fungsi menghitung heatloss dengan pendekatan resistansi
def calculateq(Ti,Tenv,Rtot):
    return (Ti-Tenv)/Rtot 


print(f'Iterasi\t Re\t\tPr\t\tNu\t\th\t\tq\t\tTs\t\tTf\t\tError')

while error > toleransi:
    
    # Kasus Dengan Insulator 1 inch dan Clading 2 mm

    # Bilangan Reynold
    Re = numberRe(Regrho(Tf),vudara,Dinsulasi,Regmiu(Tf)) 
    Nu = numberNu(Re,RegPr(Tf))
    h = hcoef(Nu,Regkf(Tf),Dinsulasi)


    # Perhitungan Resistansi
    # Inisiasi array dinamik resistansi
    resistansi = [0]
    resistansi.append(conductresistcirc(r1,r2,Regkpipa(Ti)))
    resistansi.append(conductresistcirc(r2,r3,kinsulasi))
    resistansi.append(conductresistcirc(r3,r4,kcladding))
    resistansi.append((1/(2*np.pi*h*r4)))

    resistansiTotal = np.sum(resistansi)

    heatloss.append(calculateq(Ti,Tenv,resistansiTotal))
    
    Ts = heatloss[2]*resistansi[len(resistansi)-1]+Tenv
    Tf2 = (Ts+Tenv)/2
    
    error = abs(((Tf2-Tf)/Tf)*100)

    print(f'{iter} \t {Re:.4f} \t {RegPr(Tf):.4f} \t {Nu:.4f} \t {h:.4f} \t {heatloss[2]:.4f} \t {Ts-273:.4f} \t {Tf-273:.4f} \t {error:.4f}')

    Tf = Tf2
    iter = iter +1

print("Heat loss per satuan panjang dari pipa adalah sebesar :",heatloss[2], "Watt/m")

```

### Variasi Tebal Insulator

```python

#Tebakan awal 
Tf = Tenv + 1 #Temperature Film
toleransi = 0.00001

error = 1
iter= 1

# Tebal Insulator
rins = np.arange(InchtoM(1),InchtoM(7),InchtoM(1))


# Fungsi menghitung Resistansi konduktif silinder
def conductresistcirc(rin,rout,k):
    return (np.log(rout/rin))/(2*np.pi*k)

# Fungsi menghitung Reynold Number
def numberRe(rho,v,D,miu):
    return (rho*v*D)/(miu) 

# Fungsi menghitung Nusselt Number dengan pendekatan Churchill and Bersnstein Equation
def numberNu(Re,Pr):
    return  0.3 + ((0.62*(Re**(1/2))*(Pr**(1/3))) / ((1+(0.4/Pr)**(2/3))**(1/4))) * ((1+(Re/282000)**(5/6))**(4/5))

def hcoef(Nu,kf,D):
    return (Nu*kf)/D

# Fungsi menghitung heatloss dengan pendekatan resistansi
def calculateq(Ti,Tenv,Rtot):
    return (Ti-Tenv)/Rtot 


for i in range(0,6):
    print("Tebal insulator " ,rins[i])
    while error > toleransi:
        
        # Kasus Dengan Tebal insulator bervariasi
        r3 = rins[i] + r2

        # Bilangan Reynold
        Re = numberRe(Regrho(Tf),vudara,Dinsulasi,Regmiu(Tf)) 
        Nu = numberNu(Re,RegPr(Tf))
        h = hcoef(Nu,Regkf(Tf),Dinsulasi)


        # Perhitungan Resistansi
        # Inisiasi array dinamik resistansi
        resistansi = [0]
        resistansi.append(conductresistcirc(r1,r2,Regkpipa(Ti)))
        resistansi.append(conductresistcirc(r2,r3,kinsulasi))
        resistansi.append(conductresistcirc(r3,r4,kcladding))
        resistansi.append((1/(2*np.pi*h*r4)))

        resistansiTotal = np.sum(resistansi)

        heatloss.append(calculateq(Ti,Tenv,resistansiTotal))
        
        Ts = heatloss[i+2]*resistansi[len(resistansi)-1]+Tenv
        Tf2 = (Ts+Tenv)/2
        
        error = abs(((Tf2-Tf)/Tf)*100)

        Tf = Tf2
        iter = iter +1

    print("Heat loss per satuan panjang dari pipa adalah sebesar :",heatloss[i+2], "Watt/m")
    print(" ")

```

```python
rins = np.concatenate(([0],rins))
heatlosses = np.zeros(len(rins))
for i in range(1,len(rins)+1):
    heatlosses[i-1] = heatloss[i]

# Plot Variasi Tebal Insulator terhadap Heatloss
plt.figure(figsize = (8,6))
plt.xlabel("Tebal Insulator (m)", size = 14)
plt.ylabel("Heat Loss (Watt/m)", size = 14)
plt.plot(rins,heatlosses, label = 'Variasi Tebal Insulator terhadap Heatloss',color = "green", linewidth = 3)
plt.title("Variasi Tebal Insulator terhadap Heatloss",fontweight= "bold")
plt.grid('True')
plt.show()
```
